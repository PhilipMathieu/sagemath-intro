```python
import numpy as np
import matplotlib.pyplot as plt
import warnings

# This is a hacky way to avoid annoying numpy warnings later on
warnings.filterwarnings('ignore')
```

```python
# Use numpy matrix operations to calculate the relative frequency
def relative_frequency(ints, n):
    return np.divide(
        np.cumsum(ints == n), # total number of occurrences of n up to that point
        np.arange(len(ints))) # divided by the number of integers generated up to that point
```

```python
# Set parameters
TRIALS = 100000
N = [5,13,23]

# Generate random ints
seed = 12345
rng = np.random.default_rng(seed)  # can be called without a seed
ints = rng.integers(1, 100, endpoint=True, size=(TRIALS,))
print(ints[0:10])

# Plot relative frequency
for n in N:
    plt.plot(np.arange(0,TRIALS), relative_frequency(ints, n), label=str(n))

# Format and plot
plt.title("Relative Frequency of "+str(N))
plt.xlim(0,TRIALS)
plt.xlabel("Trial Number")
plt.ylim(0,0.02)
plt.ylabel("Relative Frequency")
plt.grid(axis='y')
plt.legend()
plt.show()
```
