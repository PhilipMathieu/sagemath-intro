# Probability

Let's start by using a python function to define a mathematical function, specifically, a uniform probability density function. From the [probability and statistics cookbook](http://statistics.zone/) (or any other reference) we define that distribution as:
$$f_X(x)=\frac{I(a\leq x \leq b)}{b-a+1}$$
where $a$ and $b$ are the lower and upper ends of the uniform distribution, respectively.

````python
def uniform_pdf(x,a,b):
    return piecewise([
            ((-oo, a), 0),
            ([a, b], 1/(b-a)),
            ((b,oo), 0)],
            var=x)
    
````

Note that we are using sagemath to declare a piecewise function. This means that to actually use the function, we will need to supply values for ```a``` and ```b```:

````python
var('x')
uniform_pdf_0_1 = uniform_pdf(x,0,1)
````

We can easily plot the resulting function:

````python
plot(uniform_pdf_0_1, (x,-1,2))
````

and integrate it to find the CDF:

````python
uniform_cdf_0_1 = integrate(uniform_pdf_0_1, x)
print(uniform_cdf_0_1)
````

Last, we can plot the CDF, and even plot both functions together:

````python
plot(uniform_cdf_0_1, (x,-1,2), color='red')
````
